$$

$$
\frac{\partial f}{\partial z} = 1
$$

$$
\nabla f(x,y,z) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right) = (1,1, 1)

```python
#1 Implementation of the function f(x,y,z)
def function_OG(x, y, z):
    return np.exp(x**2) + y**y + np.exp(x*y) + z * np.cos(x)

#2 Calculate the numerical approximations of the partial derivatives using small pertubations
def numerical_approx(F, x, y, z, h=1e-5):
    #pertubation value of 1e-5 was chosen
    DF_Dx = (F(x+h, y, z) - F(x, y, z)) / h
    DF_Dy = (F(x, y+h, z) - F(x, y, z)) / h
    DF_Dz = (F(x, y, z+h) - F(x, y, z)) / h
    
#3 Comparison of analytical results from Q1a with numerical approximations
    return np.round(DF_Dx, 1), np.round(DF_Dy, 1), np.round(DF_Dz, 1)
    #Function needs to be rounded since gradient values are given with multiple significant digits

result = function_OG(0, 1, 1)
print(f"The result of the function evaluation at (0, 1, 1) is {result}")

x = 0.0
y = 1.0
z = 1.0

Analytical_DF_Dx = 1.0
Analytical_DF_Dy = 1.0
Analytical_DF_Dz = 1.0

Numerical_DF_Dx, Numerical_DF_Dy, Numerical_DF_Dz = numerical_approx(function_OG, x, y, z)

#4 Results of the function evaluation at the point (0, 1, 1)
print(f"The analytical partial derivative with respect to x would be {Analytical_DF_Dx}")
print(f"The numerical partial derivative with respect to x would be {Numerical_DF_Dx}")

print(f"The analytical partial derivative with respect to y would be {Analytical_DF_Dy}")
print(f"The numerical partial derivative with respect to y would be {Numerical_DF_Dy}")

print(f"The analytical partial derivative with respect to z would be {Numerical_DF_Dz}")
print(f"The numerical partial derivative with respect to z would be {Analytical_DF_Dz}")



```

```python
"""
Q3a (2 marks)
Preprocess the data:
- Use the clean dataset from Q2a (with missing values removed)
- Split into training (80%) and test (20%) sets using random_state=42
- Print first 5 rows of training data with their species labels
- Print shapes of both datasets
- Show number of samples per species in each split
"""

# The clean dataset
features = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
labels = penguins[['species']]

A = features
B = labels 

#Spliting the data
Atest, Atrain, Btest, Btrain = train_test_split(A, B, test_size=0.20, random_state=42)

# Printing the first 5 rows of training data with their species labels
print(" The first 5 rows of training data with the corresponding species labels:")
print(Atrain.head())
print(Btrain.head())

# Printing the shapes of the training and test datasets
print("\nShapes:")
print("Atrain:", Atrain.shape, "Btrain:", Btrain.shape)
print("Btest:", Atest.shape, "Btest:", Btest.shape)

# Show number of samples per species in each split
print("\nSpecies counts in training set:")
print(Btrain.value_counts())
print("\nSpecies counts in test set:")
print(Btest.value_counts())

```

```python
"""
Q3b (3 marks)
Train and evaluate a Decision Tree:
1. Create a Decision Tree with default parameters (random_state=42)
2. Evaluate the model:
   - Fit on training data (numerical features only)
   - Print training and test accuracy score for this tree
   - Perform 10-fold cross-validation
   - Print mean and std of cross-validation scores and build cv_scores histogram. 
   - What does the histogram show?
"""

from sklearn.model_selection import cross_val_score
tree = DecisionTreeClassifier(random_state=42)
tree.fit(Atrain, Btrain)

# Print training and test accuracy score 
training = tree.score(Atrain, Btrain)
testing = tree.score(Atest, Btest)
print('The Decision Tree Classifier :');print('The training accuracy would be {:.0f}%'.format(training * 100))
print('The test accuracy would be {:.0f}%'.format(testing * 100))

# Performing the 10-fold cross
cv_scores = cross_val_score(tree, A, B, cv=10)
print("\n10-Fold Cross-Validation Scores:")
print(cv_scores)
print("Mean CV Accuracy: {:.2f}%".format(np.mean(cv_scores) * 100))
print("Std of CV Accuracy: {:.2f}%".format(np.std(cv_scores) * 100))

# Build a histogram of cross-validation scores
plt.figure(figsize=(6,4))
plt.hist(cv_scores, bins=10, color='salmon', edgecolor='black')
plt.title("Histogram of 10-Fold Cross-Validation Scores")
plt.xlabel("Accuracy")
plt.ylabel("Frequency")

```

Interpretation: 
The decision tree model has a training accuracy of 100.00% and a test accuracy of 92%.The 10-fold cross-validation accuracy scores range from 94% to 100.00% with results in a mean of 97% and a standard deviation of 1.86%. The histogram shows the spread of the accuracy scores over the 10 folds. A low narrow spread shows that there is a consistent performance across folds.

```python
"""
Q4a (2 marks)
Prepare data for sex prediction:
- Use the clean dataset (no missing values) from Section 2
- Convert species to numeric values.

Note: Unlike the example in the tutorial, we will use a more straightforward method, LabelEncoder().

For this task, you just need to instantiate it, use the fit_transform method on the "species" column,
and reassign or add the column. 
See documentation for details:
https://scikit-learn.org/dev/modules/generated/sklearn.preprocessing.LabelEncoder.html

- Create new feature/label split using "sex" as target
- Scale features using StandardScaler().fit_transform(features) as in Tutorial
See documentation: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

- Split data into training (80%) and test (20%) sets
- Train these classifiers:
  * Logistic Regression (solver='lbfgs')
  * Decision Tree (max_depth=3)
  * KNN (n_neighbors=5)
  * SVM (kernel="linear", C=0.025)
- Compare models using (Providing values for each would be enough):
  * Training and test accuracy
  * 10-fold cross-validation scores


"""

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Step 1: Prepare features by encoding 'species' as numeric
label_encoder = LabelEncoder()
filtered_penguins['species_encoded'] = label_encoder.fit_transform(filtered_penguins['species'])

# Step 2: Create features and labels (using 'sex' as target)
features = filtered_penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species_encoded']]
labels = filtered_penguins['sex']

# Step 3: Scale features using StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 4: Split data into training (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(scaled_features, labels, test_size=0.2, random_state=42)

# Step 5: Train Classifiers
# Logistic Regression
log_reg = LogisticRegression(solver='lbfgs', max_iter=1000)
log_reg.fit(X_train, y_train)

# Decision Tree
dt_clf = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_clf.fit(X_train, y_train)

# K-Nearest Neighbors
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)

# Support Vector Machine
svm_clf = SVC(kernel='linear', C=0.025, random_state=42)
svm_clf.fit(X_train, y_train)

# Step 6: Evaluate Models (Training and Test Accuracy)
models = {
    "Logistic Regression": log_reg,
    "Decision Tree": dt_clf,
    "K-Nearest Neighbors": knn_clf,
    "Support Vector Machine": svm_clf
}

from sklearn.model_selection import cross_val_score

for model_name, model in models.items():
    
    train_accuracy = model.score(X_train, y_train)
    test_accuracy = model.score(X_test, y_test)
    cv_scores = cross_val_score(model, scaled_features, labels, cv=10)
    
    print(f"{model_name}:")
    print(f"  Training Accuracy: {train_accuracy:.3f}")
    print(f"  Test Accuracy: {test_accuracy:.3f}")
    print(f"  10-Fold Cross-Validation Score: {cv_scores.mean():.3f}\n")
```

```python
"""
Q4b (2 marks)
Use best model to predict missing sex:
- Make sure to create new variable for subset to work with and copy your variable with missing records into it,
using new_var = your_variable.copy().
- Print records with missing sex values from section 2.
- Remove records that have missing values other than sex using dropna: df.dropna(subset=['column_name'],inplace=True).
- Scale features using StandardScaler().fit_transform(features) as in Q3
- Create new features/labels variables for the new dataset. 

- Select best classifier based on test performance from Q4a

- For each record in missing_sex dataset, using best_model.predict(features) and best_model.predict_proba(features)
- Add missing values to the clean dataset and make sure there are no missing values.
"""

# Step 1: Create a copy of the dataset with missing 'sex' values
missing_sex_data = filtered_penguins[filtered_penguins['sex'].isna()].copy()

# Step 2: Check if there are any missing 'sex' values
if missing_sex_data.empty:
    print("No missing 'sex' values found. No predictions needed.")
else:
    # Step 3: Print records with missing 'sex' values
    print("Records with missing 'sex' values:")
    print(missing_sex_data)

    # Step 4: Remove records with missing values in columns other than 'sex'
    clean_penguins = filtered_penguins.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species'])

    # Step 5: Encode 'species' as numeric for both datasets
    label_encoder = LabelEncoder()
    clean_penguins['species_encoded'] = label_encoder.fit_transform(clean_penguins['species'])

    # Ensure all species in missing_sex_data exist in the encoder's classes
    missing_sex_data = missing_sex_data[missing_sex_data['species'].isin(label_encoder.classes_)]
    missing_sex_data['species_encoded'] = label_encoder.transform(missing_sex_data['species'])

    # Step 6: Prepare features for both datasets
    features_clean = clean_penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species_encoded']]
    features_missing = missing_sex_data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species_encoded']]

    # Step 7: Scale features using StandardScaler
    scaler = StandardScaler()
    scaled_features_clean = scaler.fit_transform(features_clean)
    
    if not features_missing.empty:
        scaled_features_missing = scaler.transform(features_missing)

        # Step 8: Train the best model (Assuming Logistic Regression was the best from Q4a)
        labels_clean = clean_penguins['sex']
        best_model = LogisticRegression(solver='lbfgs', max_iter=1000)
        best_model.fit(scaled_features_clean, labels_clean)

        # Step 9: Predict missing 'sex' values
        predicted_sex = best_model.predict(scaled_features_missing)

        # Step 10: Add predicted 'sex' values back into the missing_sex_data DataFrame
        missing_sex_data['sex'] = predicted_sex

        # Step 11: Integrate the predicted values back into the original dataset
        filtered_penguins.update(missing_sex_data)

    # Step 12: Verify there are no missing 'sex' values left
    print("\nFinal Dataset - Missing 'sex' values filled:")
    print(filtered_penguins.isna().sum())

```
